### Pertanyaan 1:
Bagaimana peningkatan polusi di kota-kota besar?

Bagian ini adalah tampilan chart dari pertanyaan 1 yang sudah divisualisasikan agar mempermudah proses pengambilan keputusan

Konsentrasi CO pada kota Aotizhongxin

```python
df = pd.DataFrame(aoti)

# Mengelompokkan data berdasarkan tahun dan menghitung rata-rata CO per tahun
average_co_per_year = df.groupby('year')['CO'].mean()

# Membuat plot untuk data rata-rata CO per tahun
plt.figure(figsize=(10, 6))
average_co_per_year.plot(kind='bar', color='skyblue')
plt.title('Rata-rata CO per Tahun pada Kota Aotizhongxin')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata CO')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
